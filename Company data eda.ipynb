```python
import re

def convert_funding(funding_str):
    if "B" in funding_str:
        return float(re.search(r'\d+', funding_str).group())
    elif "M" in funding_str:
        return float(re.search(r'\d+', funding_str).group()) * 0.001
    else:
        return "Not found"



comp_data['Funding_Num'] = comp_data['Funding'].apply(lambda x: convert_funding(str(x)))
```

```python
df = df[df['Funding_Num']!="Not found"]
```

```python
country_funding_valuation = df.groupby('Continent').agg({'Funding_Num': 'sum', 'Valuation_Num': 'sum'})
country_funding_valuation.plot(kind='bar', color=['lightblue', 'lightgreen'])
plt.title('Country-wise Funding vs. Valuation')
plt.ylabel('Amount ($B)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.legend(['Funding', 'Valuation'])
plt.show()
```

```python
industry_trend = comp_data.groupby(['Industry', 'Year Joined']).size().unstack(fill_value=0)
industry_trend.plot(kind='bar', stacked=True, colormap='Paired')
plt.title('Industry Trends over Time')
plt.ylabel('Number of Companies')
plt.xlabel('Year Joined')
plt.xticks(rotation=90)
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
geo_industry_analysis = comp_data.groupby(['Continent', 'Industry']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 6))
geo_industry_analysis.plot(kind='bar', stacked=True, colormap='Paired')
plt.xlabel('Continent')
plt.ylabel('Number of Companies')
plt.title('Geographical Analysis by Industry')
plt.grid(axis='y')
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
funding_patterns = df.groupby('Year Joined')['Funding_Num'].sum()
plt.figure(figsize=(10, 6))
plt.plot(funding_patterns.index, funding_patterns.values, marker='o')
plt.xlabel('Year Joined')
plt.ylabel('Funding (Billion)')
plt.title('Funding Patterns Over the Years')
plt.grid(True)
plt.show()
```

```python
valuation_trends = comp_data.groupby('Year Joined')['Valuation_Num'].sum()
plt.figure(figsize=(10, 6))
plt.plot(valuation_trends.index, valuation_trends.values, marker='o')
plt.xlabel('Year Joined')
plt.ylabel('Valuation (Billion)')
plt.title('Valuation Trends Over the Years')
plt.grid(True)
plt.show()
```

```python
investors = comp_data['Select Investors'].str.split(', ', expand=True).stack()
top_investors = investors.value_counts().head(5)
plt.figure(figsize=(10, 6))
top_investors.plot(kind='bar', color='skyblue')
plt.xlabel('Investor')
plt.ylabel('Number of Investments')
plt.title('Top 5 Investors')
plt.grid(axis='y')
plt.show()

print("Top 5 Investors Insights:")
print(top_investors)
```

```python
industry_valuation = comp_data.groupby('Industry')['Valuation_Num'].sum()
top_industries = industry_valuation.nlargest(5)
plt.figure(figsize=(10, 6))
top_industries.plot(kind='bar', color='lightcoral')
plt.xlabel('Industry')
plt.ylabel('Total Valuation')
plt.title('Top 5 Most Successful Industries')
plt.grid(axis='y')
plt.show()
print("Top 5 Most Successful Industries Insights:")
print(top_industries)
```
